```python
import os
import pandas as pd
import requests
from fake_useragent import UserAgent
from bs4 import BeautifulSoup
import numpy as np
```

```python
headers = {'User-Agent':fa.random}
city_to_code = {
    '台北':'977abb69-413a-4ccf-a109-0272c24fd490',
    '新竹':'a7a04c89-900b-4798-95a3-c01c455622f4'
}
url = "https://m.thsrc.com.tw/tw/TimeTable/SearchResult"
form_data = {
    "startStation": city_to_code['台北'],
    "endStation": city_to_code['新竹'],
    "theDay": "2019/02/15",
    "timeSelect": "18:30",
    "waySelect": "DepartureInMandarin"
}
```

```python
res = requests.post(url,headers=headers,data=form_data)
soup = BeautifulSoup(res.text,'lxml')
```

```python
def remove_space(string):
    temp_str="".join(str(string).splitlines())
    temp_str = temp_str.replace(" ","")
    return temp_str
```

```python
title=[]
for row in soup.find_all("div",{"class":"timeResultListTitle"}):
    temp_str = remove_space(row.text)
    title.append(temp_str)
```

```python
res = pd.DataFrame()

car_num = soup.find_all("a",{"class":"ui-block-a"})
deptime = soup.find_all("a",{"class":"ui-block-b"})
free_num = soup.find_all("a",{"class":"ui-block-c"})

size = len(car_num)
for i in range(size):
    car_temp = remove_space(car_num[i].text)
    deptime_temp = remove_space(deptime[i].text)
    free_num_temp = remove_space(free_num[i].text)
    row=pd.DataFrame([car_temp,deptime_temp,free_num_temp])
    res=res.append([[car_temp,deptime_temp,free_num_temp]])
res.columns = title
res = res.reset_index(drop=True)
res
```
